```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filmes_uri = 'dadosoficina03.txt'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

```python
filmes_orcamento = filmes.sort_values(by='Budget_M',ascending=False)
plt.figure(figsize=(20,20))
sns.set_style('whitegrid')
sns.barplot(x='Budget_M',y='Film',data=filmes_orcamento,palette='pastel')
plt.title('Filmes por oramento')
plt.show()
```

```python
#Filme com maior oramento
filmes_orcamento[filmes_orcamento['Budget_M']==filmes_orcamento['Budget_M'].max()]
```

```python
#Filme com menor oramento
filmes_orcamento[filmes_orcamento['Budget_M']==filmes_orcamento['Budget_M'].min()]
```

```python
filmes_bilheteria = filmes.sort_values(by='Worldwide_Gross_M',ascending=False)
plt.figure(figsize=(20,20))
sns.set_style('whitegrid')
sns.barplot(x='Worldwide_Gross_M',y='Film',data=filmes_bilheteria,palette='pastel')
plt.title('Filmes por bilheteria')
plt.show()
```

```python
#Filme com maior bilheteria
filmes_bilheteria[filmes_bilheteria['Worldwide_Gross_M']==filmes_bilheteria['Worldwide_Gross_M'].max()]
```

```python
#Filme com menor bilheteria
filmes_bilheteria[filmes_bilheteria['Worldwide_Gross_M']==filmes_bilheteria['Worldwide_Gross_M'].min()]
```

```python
#Quantidade de filmes por gnero em cada ano
plt.figure(figsize=(20,5))
sns.set(style="whitegrid")
sns.countplot(x='Year', hue='Genre', data=filmes, palette='pastel')
plt.title('Quantidade de filmes por gnero em cada ano')
plt.show()
```

```python
filmes_acao_animacao = filmes.query('Genre == "Action" | Genre == "Animation"')
```

```python
plt.figure(figsize=(20, 5))
sns.set(style = 'whitegrid')
filmes_acao_animacao
sns.countplot(x = 'Year',hue = 'Genre', palette = 'pastel', data = filmes_acao_animacao);
plt.title('Quantidade de filmes dos gneros Ao e Animao em cada ano')
plt.show()
```

```python
#Mdia de oramento de filmes por gnero em cada ano
plt.figure(figsize=(20,5))
sns.set(style="whitegrid")
sns.barplot(x='Year', y='Budget_M', hue='Genre', data=filmes, palette='pastel')
plt.title('Mdia de oramento dos filmes por gnero em cada ano')
plt.show()
```
