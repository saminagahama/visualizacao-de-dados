```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('dadosoficina07.txt')
df.head()
```

```python
plt.figure(figsize=(20,5))
plt.title('Vendas mensais de shampoo durante um perodo de 3 anos')
sns.set(style="whitegrid")
sns.lineplot(x="Month", y="Sales", ci = None, data=df)
plt.show()
```
