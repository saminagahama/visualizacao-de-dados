```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Distribuição do valor da passagem de acordo com a cidade na qual o passageiro embarcou')
sns.swarmplot(x="embark_town", y="fare", data=titanic)
plt.show()
```

```python
sns.catplot(x="embark_town", y="fare", col = 'survived', hue = "sex", kind = 'violin', data=titanic)
```

```python
#Pelos gráficos, nota-se que a maioria dos sobreviventes embarcaram na cidade de Cherbourg
#Aqueles que pagaram um valor mais alto pela passagem tiveram maior probabilidade de sobreviver
```
