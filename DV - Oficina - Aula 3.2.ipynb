```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
sns.lmplot(x="mpg", y="acceleration", hue = "origin", data=df)
plt.show()
```

```python
# Há uma baixa correlação positiva entre a aceleração e o consumo de combustível em todas as 3 origens
```

```python
sns.relplot(x="mpg", y="acceleration", col = "model_year", col_wrap=2, data=df)
```

```python
# Cada ano apresenta um gráfico diferente, portanto não há como concluir pela correlação entre aceleração 
# e consumo de combustível neste caso
```
